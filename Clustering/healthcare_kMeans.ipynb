```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, accuracy_score

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('healthcare-dataset-stroke-data.csv')
df
```

```python
df = df.drop("id",axis=1)
```

```python
df['bmi'] = df['bmi'].replace(np.nan, df['bmi'].mean())
```

```python
df.drop(df.loc[df['gender']=="Other"].index, inplace=True)    
```

```python
df.replace({'gender':{'Female':0,'Male':1}, 'ever_married':{'No':0,'Yes':1},'smoking_status':{'Y':1,'N':0},"Residence_type":{"Urban":0,"Rural":1}}, inplace=True)
```

```python
df.info()
```

```python
categorical_features = ['work_type','smoking_status']
dummies = pd.get_dummies(df[categorical_features], drop_first=True)
```

```python
df = pd.concat([df,dummies], axis=1, ignore_index=False)
```

```python
df = df.drop(['work_type','smoking_status'],axis=1) 
```

```python
df
```

```python
df.info()
```

```python
from sklearn.cluster import KMeans

model = KMeans(random_state=22)
model.fit(df)
```

```python
### parametri che ha utilizzato il modello in automatico
model.get_params()
```

```python
### centri dei cluster
model.inertia_
```

```python
inertia_list = list()
for n in range(2, 10):
    model = KMeans(n_clusters=n)
    model.fit(df)
    inertia_list += [model.inertia_]

plt.scatter(range(2, 10), inertia_list)
plt.plot(range(2, 10), inertia_list)
plt.title('ELBOW PLOT', fontsize=14)
plt.xlabel('n_clusters', fontsize=14)
plt.ylabel('inertia', fontsize=14)
plt.show()
```

```python
from yellowbrick.cluster import KElbowVisualizer

visualizer = KElbowVisualizer(model, k=(2,10))
visualizer.fit(df)        # Fit the data to the visualizer
visualizer.show()
```

```python
from sklearn.metrics import silhouette_score

for n in range(2,7):

    model = KMeans(n_clusters=n)
    model.fit(df)
    labels = model.labels_
    print('Avg Sihlouette Score (k={}): {} '.format(n,silhouette_score(df, labels)))
```

```python
from yellowbrick.cluster import SilhouetteVisualizer

for n in range(2,7):

    model = KMeans(n_clusters=n)
    visualizer = SilhouetteVisualizer(model, colors='yellowbrick')
    visualizer.fit(df)
    print('\nAvg Sihlouette Score: ', visualizer.silhouette_score_)
    plt.show()
```
