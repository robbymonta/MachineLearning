### CORRELAZIONI

```python
df.corr()
```

```python
corr_matrix = df.corr()

plt.figure(figsize=(10,9))
sns.heatmap(corr_matrix)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

```python
sale_corr = corr_matrix["selling_price"].sort_values(ascending=False)

plt.figure(figsize=(10,7))
sns.barplot(x=sale_corr.index, y=sale_corr)
plt.xticks(rotation=90, fontsize=12)
plt.show()
```

```python
# plottare la matrice di correlazione con sns.heatmap() e individuare le variabili più correlate con selling_price, oltre che quelle che sembrano essere collineari tra loro
most_corr_vars = sale_corr.abs().iloc[:8].index
most_corr_matrix = df[most_corr_vars].corr()
plt.figure(figsize=(10,9))

hm = sns.heatmap(most_corr_matrix, cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size': 10})
plt.show()
### ...........

plt.show()
```
